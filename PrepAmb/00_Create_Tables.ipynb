```python
# Databricks notebook source

# ============================================================
# Create Retail Lakehouse Tables
# Catalog: adbsmartdatatdp
# ============================================================

catalog = "adbsmartdatatdp"

# COMMAND ----------

spark.sql(f"USE CATALOG {catalog}")

# ============================================================
# BRONZE
# ============================================================

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.bronze
""")

# Customers Raw
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.bronze.customers_raw
(
    customer_id INT,
    first_name STRING,
    last_name STRING,
    email STRING,
    phone STRING,
    city STRING,
    registration_date DATE,
    ingestion_timestamp TIMESTAMP
)
USING DELTA
""")

# Products Raw
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.bronze.products_raw
(
    product_id INT,
    product_name STRING,
    category STRING,
    brand STRING,
    price DOUBLE,
    stock INT,
    ingestion_timestamp TIMESTAMP
)
USING DELTA
""")

# Sales Raw
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.bronze.sales_raw
(
    sale_id INT,
    customer_id INT,
    product_id INT,
    branch_id INT,
    employee_id INT,
    quantity INT,
    unit_price DOUBLE,
    sale_date DATE,
    ingestion_timestamp TIMESTAMP
)
USING DELTA
""")

# Branches Raw
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.bronze.branches_raw
(
    branch_id INT,
    branch_name STRING,
    city STRING,
    region STRING,
    ingestion_timestamp TIMESTAMP
)
USING DELTA
""")

# Employees Raw
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.bronze.employees_raw
(
    employee_id INT,
    first_name STRING,
    last_name STRING,
    position STRING,
    branch_id INT,
    hire_date DATE,
    ingestion_timestamp TIMESTAMP
)
USING DELTA
""")

# ============================================================
# SILVER
# ============================================================

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.silver
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.silver.customers
(
    customer_id INT,
    full_name STRING,
    email STRING,
    phone STRING,
    city STRING,
    registration_date DATE
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.silver.products
(
    product_id INT,
    product_name STRING,
    category STRING,
    brand STRING,
    price DECIMAL(10,2),
    stock INT
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.silver.sales
(
    sale_id INT,
    customer_id INT,
    product_id INT,
    branch_id INT,
    employee_id INT,
    quantity INT,
    unit_price DECIMAL(10,2),
    subtotal DECIMAL(10,2),
    total DECIMAL(10,2),
    sale_date DATE
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.silver.branches
(
    branch_id INT,
    branch_name STRING,
    city STRING,
    region STRING
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.silver.employees
(
    employee_id INT,
    full_name STRING,
    position STRING,
    branch_id INT,
    hire_date DATE
)
USING DELTA
""")

# ============================================================
# GOLD
# ============================================================

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.gold
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.gold.sales_summary
(
    sale_date DATE,
    total_sales DECIMAL(18,2),
    total_orders BIGINT
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.gold.sales_by_city
(
    city STRING,
    total_sales DECIMAL(18,2)
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.gold.sales_by_category
(
    category STRING,
    total_sales DECIMAL(18,2)
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.gold.top_customers
(
    customer_id INT,
    full_name STRING,
    total_amount DECIMAL(18,2)
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.gold.top_products
(
    product_name STRING,
    units_sold BIGINT
)
USING DELTA
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.gold.kpi_dashboard
(
    total_revenue DECIMAL(18,2),
    average_ticket DECIMAL(18,2),
    total_customers BIGINT,
    total_products BIGINT
)
USING DELTA
""")

print("========================================")
print("Lakehouse structure created successfully")
print("Catalog :", catalog)
print("Schemas : bronze, silver, gold")
print("Tables  : 16")
print("========================================")
```
