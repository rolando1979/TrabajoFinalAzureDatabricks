```python
# Databricks notebook source

# MAGIC %md
# MAGIC # Data Access Report
# MAGIC
# MAGIC Este notebook genera un reporte de los permisos existentes
# MAGIC sobre el catálogo, esquemas y tablas del proyecto.

# COMMAND ----------

catalog = "adbsmartdatatdp"

# COMMAND ----------

print("="*70)
print("CATALOG")
print("="*70)

display(spark.sql(f"""
SHOW GRANTS ON CATALOG {catalog}
"""))

# COMMAND ----------

for schema in ["bronze","silver","gold"]:

    print("="*70)
    print(f"SCHEMA: {schema}")
    print("="*70)

    display(spark.sql(f"""
    SHOW GRANTS ON SCHEMA {catalog}.{schema}
    """))

# COMMAND ----------

tables = [
    "gold.sales_summary",
    "gold.sales_by_city",
    "gold.sales_by_category",
    "gold.top_customers",
    "gold.top_products",
    "gold.kpi_dashboard"
]

for table in tables:

    print("="*70)
    print(table)
    print("="*70)

    display(spark.sql(f"""
    SHOW GRANTS ON TABLE {catalog}.{table}
    """))
```
