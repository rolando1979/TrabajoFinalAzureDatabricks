```python
# Databricks notebook source

# ============================================================
# Drop Retail Lakehouse
# Catalog: adbsmartdatatdp
# ============================================================

catalog = "adbsmartdatatdp"

print("="*60)
print("STARTING LAKEHOUSE CLEANUP")
print("="*60)

# ============================================================
# DROP GOLD TABLES
# ============================================================

print("Dropping Gold tables...")

gold_tables = [
    "kpi_dashboard",
    "top_products",
    "top_customers",
    "sales_by_category",
    "sales_by_city",
    "sales_summary"
]

for table in gold_tables:
    spark.sql(f"DROP TABLE IF EXISTS {catalog}.gold.{table}")

print("Gold tables removed.")

# ============================================================
# DROP SILVER TABLES
# ============================================================

print("Dropping Silver tables...")

silver_tables = [
    "sales",
    "customers",
    "products",
    "branches",
    "employees"
]

for table in silver_tables:
    spark.sql(f"DROP TABLE IF EXISTS {catalog}.silver.{table}")

print("Silver tables removed.")

# ============================================================
# DROP BRONZE TABLES
# ============================================================

print("Dropping Bronze tables...")

bronze_tables = [
    "sales_raw",
    "customers_raw",
    "products_raw",
    "branches_raw",
    "employees_raw"
]

for table in bronze_tables:
    spark.sql(f"DROP TABLE IF EXISTS {catalog}.bronze.{table}")

print("Bronze tables removed.")

# ============================================================
# DROP SCHEMAS
# ============================================================

print("Dropping schemas...")

schemas = ["gold", "silver", "bronze"]

for schema in schemas:
    spark.sql(f"DROP SCHEMA IF EXISTS {catalog}.{schema} CASCADE")

print("Schemas removed.")

# ============================================================
# DROP CATALOG
# ============================================================

print("Dropping catalog...")

spark.sql(f"DROP CATALOG IF EXISTS {catalog} CASCADE")

print("Catalog removed.")

# ============================================================
# END
# ============================================================

print("="*60)
print("LAKEHOUSE REMOVED SUCCESSFULLY")
print("="*60)
```
